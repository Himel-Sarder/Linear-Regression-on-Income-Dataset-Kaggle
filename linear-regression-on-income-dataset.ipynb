```python
X = df.iloc[:,1:2]
y = df.iloc[:,-1]
```


## X: Selects the second column (income) as the feature.

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# y_test: actual values
# y_pred: predicted values

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# Calculate R-squared (RÂ²)
r2 = r2_score(y_test, y_pred)
print("R-squared:", r2)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
# Calculate metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

These values indicate that your model has drastically improved! Let's break down what each metric means in this context:

1. **Mean Absolute Error (MAE) = 3.51e-16**: 
   - This is extremely close to zero, which suggests that the predictions are nearly identical to the actual values. A value this small indicates that your model is performing very well.

2. **Mean Squared Error (MSE) = 1.87e-31**: 
   - This is also very close to zero. Since MSE squares the errors, the fact that it's this small further supports that your model's predictions are almost perfect.

3. **R-squared = 1.0**: 
   - This is the best possible value for R-squared, indicating that your model explains 100% of the variance in the data. In other words, your model perfectly fits the data.

### Conclusion:
These values suggest that your model is performing excellently and that the predictions are almost identical to the actual values. This could happen if:
- Your model is overfitting (especially if you used a very small or simple dataset).
- The data is almost perfectly linear or the features and target are highly correlated.

### Recommendations:
- **Check for Overfitting**: If you have a small dataset, try testing your model on new, unseen data to ensure it generalizes well.
- **Cross-Validation**: If possible, use cross-validation to check the model's performance on different splits of the data.
